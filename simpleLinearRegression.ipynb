SIMPLE LINEAR REGRESSION

Importing the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Importing the dataset

```python
dataset=pd.read_csv('/content/Salary_Data.csv')
dataset.head()
```

```python
X=dataset.iloc[:,:-1].values
y=dataset.iloc[:,-1].values
```

Splitting the dataset into train and test

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

Training the simple linear regression model

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,y_train )
```

Predicting the test set results

```python
y_pred=regressor.predict(X_test)#y_pred is the one predicted by the model y_test is the actual one
```

Visualising the training set

```python
plt.scatter(X_train,y_train,color='red')#real values of age and salaries
plt.plot(X_train,regressor.predict(X_train),color='blue')#predicted salaries of the train set
plt.title('salary vs experience(Training set)')
plt.xlabel('Years of experience')
plt.ylabel('Salary')
plt.show()
```

```python
plt.scatter(X_test,y_test,color='red')#real values of age and salaries
plt.plot(X_train,regressor.predict(X_train),color='blue')#predicted salaries of the train set
plt.title('salary vs experience(Test set)')
plt.xlabel('Years of experience')
plt.ylabel('Salary')
plt.show()
```
